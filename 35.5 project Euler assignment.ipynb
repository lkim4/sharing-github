```python
def isPrime(n):
    if n < 2: return "Neither prime, nor composite"
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def nthPrime(n):
    numberOfPrimes = 0
    prime = 1

    while numberOfPrimes < n:
        prime += 1
        if isPrime(prime):
            numberOfPrimes += 1
    return prime

print(nthPrime(10001))
```

10. summation of primes

The sum of the primes below 10 is 2 + 3 + 5 + 7 = 17.

Find the sum of all the primes below two million.

```python
original_list=[2,3,4,5,6,7,8,9]
new_list=[]
```

```python
for i in original_list:
    x=np.arange(2,i)
    a=i % x
    if a.all() != 0:
        new_list.append(i)
print(sum(new_list))
```

```python
def sumPrimes(n):
    sum, sieve = 0, [True] * n
    for i in range(2, n):
        if sieve[i]:
            sum += i
            for i in range(i*i, n, i):
                sieve[i] = False
    return sum

print (sumPrimes(2000000))
```
