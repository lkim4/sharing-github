```python
data['average score']=(data['math score']+ data['reading score']+ data['writing score'])/3
column1=pd.DataFrame(data['gender'].reset_index(drop=True))
column2=pd.DataFrame(data['race/ethnicity'].reset_index(drop=True))
column3=pd.DataFrame(data['parental level of education'].reset_index(drop=True))
column4=pd.DataFrame(data['lunch'].reset_index(drop=True))
column5=pd.DataFrame(data['test preparation course'].reset_index(drop=True))
column6=pd.DataFrame(data['average score'].reset_index(drop=True))
new_table=column1.join([column2, column3, column4, column5, column6])
new_table
```

```python
def pass_fail():
    p_f = []
    for x in new_table['average score']:
        if x >= 70:
            p_f.append('Pass')
        else:
            p_f.append('Fail')
    return p_f
```

```python
new_table['pass/fail']=pass_fail()
new_table
```

```python
r_sample=np.random.choice(r['average score'], 100, replace=True)
p_sample=np.random.choice(p['average score'], 100, replace=True)

plt.hist(r_sample, alpha=0.5, label='male sample')
plt.hist(p_sample, alpha=0.5, label='female sample')
plt.legend(loc ='upper right')
plt.show()
```

```python
print(r_sample.mean())
print(p_sample.mean())
print(r_sample.std())
print(p_sample.std())
diff=p_sample.mean( ) -r_sample.mean()
print(diff)
```

```python
size = np.array([len(r_sample), len(p_sample)])
sd = np.array([r_sample.std(), p_sample.std()])
diff_se = (sum(sd ** 2 / size)) ** 0.5   
print(diff/diff_se)
```

```python
from scipy.stats import ttest_ind
print(ttest_ind(p_sample, r_sample, equal_var=False))
```

```python
#t-value of 1.7885398, indicates male_sample and female_sample's means are 1.7885398 standard errors apart.  
#p-value is from 0-1.  Smaller p-value: "more confidently we can conclude that there is a meaningful difference 
#between the means of the two groups in the population. "Difference between the means are due to actual difference 
#in two population sample rather then due to variance in the sample. 
#our p-value is small.  So the difference between male and female students' overall test score average
#are due to actual difference in two population. 
```
