```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Downloads/MyLA311_Service_Request_Data_2016.csv')
column1=pd.DataFrame(df['CreatedDate'].astype(str).str.slice(stop=2)).reset_index(drop=True)
column2=pd.DataFrame(df['RequestSource'].astype(str)).reset_index(drop=True)
df1=pd.DataFrame(column1.join(column2))
df1.columns=['month', 'request source']
```

```python
def request_source():
    list = []
    for x in df1['request source']:
        if (x == 'Call'):
            list.append('Call')
        elif (x == 'Mobile App'):
            list.append('Mobile App')
        elif (x == 'Twitter'):
            list.append('Twitter')
        else:
            list.append('Others')
    return list
```

```python
df1['request source']=request_source()
```

```python
data=df1['request source'].value_counts()
data.plot(kind='bar')
plt.title('request source counts for 2016')
plt.show()
```

```python
table=pd.DataFrame(df1.groupby(['month'])['request source'].value_counts())
table.columns=['request source counts']
table
```

```python
table1=pd.DataFrame(df1.groupby(['month'])['request source'].count())
table1.columns=['number of request']
table1
```

```python
table1.plot(kind='bar')
plt.title('number of request for each month of 2016')
plt.show()
```
