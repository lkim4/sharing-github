```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df3=pd.read_csv('Downloads/MyLA311_Service_Request_Data_2018.csv')
column1=pd.DataFrame(df3['CreatedDate'].astype(str).str.slice(stop=2)).reset_index(drop=True)
column2=pd.DataFrame(df3['RequestType'].astype(str)).reset_index(drop=True)
df4=pd.DataFrame(column1.join(column2))
df4.columns=['month', 'request type']
df4.sort_values(by=['month'],ascending=True)
df4
```

```python
requestTypeCounts=pd.DataFrame(df4.groupby(['month'])['request type'].value_counts().reset_index(drop=True))
```

```python
requestType2018=df4.groupby(['month','request type']).count().reset_index()
requestType2018['Counts']=requestTypeCounts['request type']
requestType2018
```

```python
trend2018=df4.groupby(df4['month'])['request type'].value_counts()
trend2018.plot(kind='bar', figsize=(25,8))
plt.title('total number of request per request type for each month')
plt.ylabel('total number of request')
plt.show()
```
