```python
import numpy as np
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
from sqlalchemy import create_engine

import warnings
warnings.filterwarnings('ignore')

postgres_user = 'dsbc_student'
postgres_pw = '7*.8G9QH21'
postgres_host = '142.93.121.174'
postgres_port = '5432'
postgres_db = 'houseprices'
```

```python
engine = create_engine('postgresql://{}:{}@{}:{}/{}'.format(postgres_user, postgres_pw, postgres_host, postgres_port, postgres_db))
houseprice_df = pd.read_sql_query('select * from houseprices',con=engine)

engine.dispose()

houseprice_df
```

```python
houseprice_df.dropna(axis='columns')
```

```python
Y = houseprice_df['saleprice']
X = houseprice_df[['overallqual','grlivarea','garagecars','garagearea','totalbsmtsf','firstflrsf']]

model = linear_model.LinearRegression()
model.fit(X, Y)
y_predict = model.predict(X)

print('coefficients:', model.coef_)
print('y-intercept:', model.intercept_)
```

```python
#equation of the best fit line to predict the saleprice:
#saleprice = -102650.9 + 2.3997e+04overallqual + 4.3123e+01grlivarea + 1.4515e+04garagecars + 1.5664e+01garagearea 
# + 2.4391e+01totalbsmtsf + 1.1186e+01firstflrsf
```

```python
#assumption one: linearity of the model in its coefficient.
#each explanatory variable have one coefficient.  
#assumption one is met.
```

```python
#assumption two: avg of error terms are zero
#because actual y values are distributed equally on either side of the predicted y-value line
```

```python
errors = Y - y_predict
```

```python
np.mean(errors)
```

```python
#avg of error terms is pretty much zero, so assumption two is met.
```

```python
#assumption three: homoscedasticity.  Error variance is about the same along the y_predict line.
```

```python
from scipy.stats import levene

lev_stats = levene(y_predict, errors)
lev_stats
```

```python
#p-value for levene test is below 0.05, meaning the null hypothesis is rejected (null hypothesis: homoscedasticity).
#assumption three is not met.
```

```python
#assumption 4: low multicollinearity
```

```python
df=houseprice_df[['saleprice','overallqual','grlivarea','garagecars','garagearea','totalbsmtsf','firstflrsf']]
df.corr()
```

```python
#they have moderate level of correlation.
#assumption 4 is not met.
```

```python
#assumption 5: no autocorrelation
```

```python
plt.plot(errors)
```

```python
from statsmodels.tsa.stattools import acf

acf_data = acf(errors)

plt.plot(acf_data[1:])     
plt.show()                
```

```python
#correlation values are low (ranging from around -0.06 to 0.06).
#assumption 5 is met
```
