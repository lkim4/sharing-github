```python
import pandas as pd
import numpy as np
import statistics
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('Downloads/MyLA311_Service_Request_Data_2019.csv')
```

```python
df['month']=df['CreatedDate'].astype(str).str[1]
```

```python
year2019_column1=pd.DataFrame(df['month'])
year2019_column1=year2019_column1.sort_values(by='month', ascending=True)
year2019_column2=pd.DataFrame(df['RequestType'])
year2019=year2019_column1.join(year2019_column2)
```

```python
change_month_dtype = pd.DataFrame(year2019.month.astype(int))
year2019 = change_month_dtype.join(year2019_column2)
year2019
```

```python
requestTypeCounts=pd.DataFrame(year2019.groupby(['month'])['RequestType'].value_counts().reset_index(drop=True))
```

```python
requestType2019=year2019.groupby(['month','RequestType']).count().reset_index()
requestType2019['Counts']=requestTypeCounts['RequestType']
requestType2019
```

```python
trend2019=year2019.groupby(year2019['month'])['RequestType'].value_counts()
trend2019.plot(kind='bar', figsize=(13,8))
plt.title('total number of request per request type for each month')
plt.ylabel('total number of request')
plt.show()
```
