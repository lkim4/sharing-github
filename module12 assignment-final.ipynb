```python
table3=pd.DataFrame(tabley.groupby(['Journal title'])['Article title'].count())
table4=pd.DataFrame(table3.sort_values(['Article title'], ascending=False))
print(table4)
```

```python
table5=pd.DataFrame(tabley.groupby(['Journal title'])['Cost'].mean())
print(table5)
```

```python
table8=table4.join(table5)
column3=pd.DataFrame(tabley.groupby(['Journal title'])['Cost'].median())
column4=pd.DataFrame(tabley.groupby(['Journal title'])['Cost'].std())
table8['cost median']=column3
table8['cost std']=column4
print(table8)
```

```python
table8.rename({'Article title':'Journal counts', 'Cost':'mean cost', 'cost median':'median cost', 'cost std':'SD cost'}, axis='columns')
```
