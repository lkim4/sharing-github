```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats.stats import ttest_ind
import seaborn as sns
```

```python
df=pd.read_csv('/Users/laurakim/Downloads/StudentsPerformance.csv')
df
```

```python
#1.Are there any differences between the genders, ethnicities, and parental level of education with respect to their performances in exams?
```

```python
df1=df.groupby(['gender']).mean().reset_index()
df1
```

```python
ttest_ind(df[df['gender']=='female'][['math score','reading score','writing score']], df[df['gender']=='male'][['math score','reading score','writing score']])
```

```python
#between female and male:
#avg math score diff: females are lower. very small p-value.  this difference is significant
#avg reading score diff: males are lower. this difference is significant.  
#avg writing score diff: males are lower. this difference is significant. 
```

```python
df2=df.groupby(['race/ethnicity']).mean()
df2
```

```python
race = df["race/ethnicity"].unique()
for x in ["math score", "reading score", "writing score"]:
    print("------------------------------------------------")
    print("Comparisons for variable: {}".format(x))
    print("------------------------------------------------")
    for i in range(0, len(race)):
        for j in range(i+1, len(race)):
            print(
                "t-test between groups {0} and {1}:".format(race[i], race[j]))
            print(ttest_ind(
                df[df["race/ethnicity"]
                           == race[i]][x],
                df[df["race/ethnicity"] == race[j]][x]
            ))
```

```python
df3=df.groupby(['parental level of education']).mean()
df3
```

```python
parental = df["parental level of education"].unique()
for y in ["math score", "reading score", "writing score"]:
    print("------------------------------------------------")
    print("Comparisons for variable: {}".format(y))
    print("------------------------------------------------")
    for i in range(0, len(parental)):
        for j in range(i+1, len(parental)):
            print(
                "t-test between groups {0} and {1}:".format(parental[i], parental[j]))
            print(ttest_ind(
                df[df["parental level of education"]
                           == parental[i]][y],
                df[df["parental level of education"] == parental[j]][x]
            ))
```
