```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('Downloads/MyLA311_Service_Request_Data_2015.csv')
column1=pd.DataFrame(df['CreatedDate'].astype(str).str.slice(stop=2)).reset_index(drop=True)
column2=pd.DataFrame(df['RequestType'].astype(str)).reset_index(drop=True)
df1=pd.DataFrame(column1.join(column2))
df1.columns=['month', 'request type']
df1.sort_values(by=['month'],ascending=True)
df1
```

```python
requestTypeCounts=pd.DataFrame(df1.groupby(['month'])['request type'].value_counts().reset_index(drop=True))
```

```python
requestType2015=df1.groupby(['month','request type']).count().reset_index()
requestType2015['Counts']=requestTypeCounts['request type']
requestType2015
```

```python
trend2015=df1.groupby(df1['month'])['request type'].value_counts()
trend2015.plot(kind='bar', figsize=(25,8))
plt.title('total number of request per request type for each month')
plt.ylabel('total number of request')
plt.show()
```
